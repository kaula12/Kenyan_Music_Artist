```python
figsize = (15, 10)
plt.figure(figsize=figsize)
sns.boxplot(data=final)
plt.title('final')
```

```python
#Outlier detection
figsize = (15, 10)
plt.figure(figsize=figsize)
sns.boxplot(data=final, showfliers = False)
plt.title('final')
```

```python
#Using Matplot

final.plot(kind='box', subplots=True, layout=(6,6), sharex=False, sharey=False, figsize=(15, 10), showfliers = False)
plt.show()
```

```python
#Feature engineering
```

```python
#Transformation
#Converting semester scores to percentage
final['G1%'] = final['G1'] * 5
final['G2%'] = final['G2'] * 5
final['G3%'] = final['G3'] * 5
final
```

```python
#Dropping columns
final = final.drop(['G1', 'G2', 'G3'], axis=1)
final
```

###Graphs

```python
#line graph

sample_data = final[['G1%', 'G2%', 'G3%']].head(10)
sample_data.plot(kind='line', figsize=(15, 10))
plt.title('Grade Percentage Per Semester')
plt.xlabel('Student')
plt.ylabel('Percentage')
plt.show()

```
