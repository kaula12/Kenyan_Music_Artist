**Visualization on a population data**

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#reading file
population_data = pd.read_excel('/content/realistic_population_census_50k.xlsx')
population_data.head()
```

```python
#Percentage of Working and Unemployed Individuals

import matplotlib.pyplot as plt
import seaborn as sns

employment_counts = population_data['Occupation'].apply(lambda x: 'Unemployed' if x in ['Unemployed', 'None'] else 'Working')
employment_percentages = employment_counts.value_counts(normalize=True) * 100

plt.figure(figsize=(6, 4))
sns.barplot(x=employment_percentages.index, y=employment_percentages.values, palette='viridis')
plt.title('Employment Status Percentage')
plt.ylabel('Percentage')
plt.xlabel('Employment Status')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

90% of the population in our data sample is working class.
Employment rate is high in that town.

```python
# Marital Status Distribution

marital_status_counts = population_data['Marital Status'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 5))
sns.barplot(x=marital_status_counts.index, y=marital_status_counts.values, palette='pastel')
plt.title('Marital Status Distribution')
plt.ylabel('Percentage')
plt.xlabel('Marital Status')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

Over 40% of the population in our data set is single.
Mortality rate in the population is low because less than 20% of the population is widowed.

```python
# Income Distribution

plt.figure(figsize=(10, 6))
sns.histplot(population_data['Income ($)'], bins=50, kde=True, color='skyblue')
plt.title('Income Distribution')
plt.xlabel('Income ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

The larger percentage of the population earns between 40,000 and 60,000.
The above visualization shows a positively skewed representation of data.


```python
# Distribution of Education Levels

education_counts = population_data['Education Level'].value_counts(normalize=True) * 100

plt.figure(figsize=(10, 5))
sns.barplot(x=education_counts.index, y=education_counts.values, palette='Set2')
plt.title('Education Level Distribution')
plt.ylabel('Percentage')
plt.xlabel('Education Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

More than 40% of the population did not go past high school level of education.
Over 90% of the population went to school.
The higher you go in education, the lower the population percentage.

```python
# Scatter Plot: Age vs. Income by Education Level

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=population_data,
    x='Age',
    y='Income ($)',
    hue='Education Level',
    alpha=0.6
)
plt.title('Scatter Plot: Age vs. Income by Education Level')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# Heatmap: Correlation Between Numerical Features

numeric_df = population_data[['Age', 'Income ($)', 'Household Size']]
corr = numeric_df.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Correlation Between Numerical Features')
plt.tight_layout()
plt.show()
```

The correlation between age and income is high.As people get older,they tend to earn more.
The correlation between household size and income is high. Larger households have higher incomes.

```python
# line Graph: Average Income by Age

avg_income_by_age = population_data.groupby('Age')['Income ($)'].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_income_by_age.index, y=avg_income_by_age.values, color='teal')
plt.title('Line Graph: Average Income by Age')
plt.xlabel('Age')
plt.ylabel('Average Income ($)')
plt.tight_layout()
plt.show()
```

Average income rises steadily from age 20 to 30 and peaks in the 40s to early 50s.
At the age of 80, income is lowest because old people are retired or do not work.
